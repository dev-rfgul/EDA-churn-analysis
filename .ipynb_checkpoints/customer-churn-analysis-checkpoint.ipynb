```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('customer-churn.csv')
df.head(10)
```

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ", "0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value ==1:
        return "yes"
    else :
        return "no"
df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

#converted the 0 and 1 of senior citizen column into yes or no 

```python
ax=sns.countplot(x ='Churn', data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of customers which had churned")
plt.show()
```

```python
plt.figure(figsize=(4,5))
gb=df.groupby("Churn").agg({'Churn':'count'})
gb
plt.pie(gb['Churn'], labels=gb.index, autopct= "%1.3f%%")
plt.title("Percentage of Churned Customers")
plt.show()
```

```python
ax=sns.countplot(x="SeniorCitizen", data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

#comparatively a greater % of customers in senior citized have churned out

```python
# Calculate the percentage
percentage_df = (
    df.groupby(["SeniorCitizen", "Churn"])
    .size()
    .reset_index(name="Count")
)
total_counts = percentage_df.groupby("SeniorCitizen")["Count"].transform("sum")
percentage_df["Percentage"] = percentage_df["Count"] / total_counts * 100

# Create a pivot table for stacked bar plot
pivot_df = percentage_df.pivot(index="SeniorCitizen", columns="Churn", values="Percentage")

# Plot the stacked bar chart
pivot_df.plot(kind="bar", stacked=True, figsize=(8, 6),)
plt.title("Churn of Senior Citizen (Stacked by Percentage)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")

# Annotate the plot with percentages
for i, bar in enumerate(plt.gca().patches):
    height = bar.get_height()
    if height > 0:
        plt.gca().text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_y() + height / 2,
            f"{height:.1f}%",
            ha="center",
            va="center",
            fontsize=10,
            color="white",
            fontweight="bold",
        )

plt.legend(title="Churn", loc="upper left")
plt.show()
```

```python
# Define the columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots
num_columns = len(columns)
fig, axes = plt.subplots(nrows=(num_columns + 2) // 3, ncols=3, figsize=(15, 12))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Loop through each column and create a count plot
for i, column in enumerate(columns):
    ax=sns.countplot(data=df, x=column, ax=axes[i], hue=df["Churn"])
    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])
                 
    axes[i].set_title(column)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```

#the majority of the customer who do not churn out tend to have services like phone service, internet service particularly the dsl and online security enabled. for servicecs like online backup ,tech support and streaming tv  churn rated are relatively  higher when these services are not used of unavailable
